```python
import pandas as pd
import numpy as np

# Lab Task 1: Creating Series and DataFrames

# 1. Creating a Series with Middle Eastern countries
middle_east_countries = pd.Series(['Saudi Arabia', 'UAE', 'Qatar', 'Bahrain', 'Kuwait'], 
                                  index=['Iran', 'Iraq', 'Syria', 'Lebanon', 'Jordan'])
print("Middle East Countries Series:\n", middle_east_countries)

# 2. Creating a Series with average height of male and female students
height_series = pd.Series([175, 162], index=['Male', 'Female'])
print("\nHeight Series:\n", height_series)

# 3. Creating a DataFrame for student details
students_df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'John', 'Mia', 'Ahmed'],
    'Age': [22, 21, 23, 20, 24],
    'Major': ['CS', 'Physics', 'Math', 'Biology', 'Chemistry']
}, index=['S1', 'S2', 'S3', 'S4', 'S5'])
print("\nStudents DataFrame:\n", students_df)

# 4. Creating a DataFrame for sales revenue
sales_df = pd.DataFrame({
    'Product A': [500, 600, 550, 700, 650, 620],
    'Product B': [400, 420, 450, 480, 500, 530],
    'Product C': [300, 320, 330, 340, 350, 360],
    'Product D': [200, 250, 230, 270, 290, 310]
}, index=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
print("\nSales DataFrame:\n", sales_df)
```

```python
# Lab Task 2: Perform Series Indexing

# Selecting row S3 using iloc
selected_row = students_df.iloc[2]
print("\nSelected Row S3:\n", selected_row)

# Selecting physics grade of S2 using loc
grades_df = pd.DataFrame({
    'Math': [85, 90, 78],
    'Physics': [88, 92, 80],
    'Chemistry': [89, 85, 83]
}, index=['S1', 'S2', 'S3'])
physics_grade_s2 = grades_df.loc['S2', 'Physics']
print("\nPhysics Grade of S2:", physics_grade_s2)

# Selecting first three games scores using iloc
players_df = pd.DataFrame({
    'Game1': [10, 15, 20, 12],
    'Game2': [14, 18, 22, 16],
    'Game3': [17, 19, 24, 20],
    'Game4': [22, 25, 30, 28],
    'Game5': [25, 29, 35, 32]
}, index=['P1', 'P2', 'P3', 'P4'])
selected_games = players_df.iloc[:, :3]
selected_p3 = players_df.loc['P3', :]
print("\nSelected First Three Games Scores:\n", selected_games)
print("\nSelected Scores of P3:\n", selected_p3)
```

```python
# Lab Task 3: Perform DataFrames Indexing

# Selecting height of student S4
heights_series = pd.Series([170, 165, 180, 175, 160, 172], 
                           index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
height_s4 = heights_series.iloc[3]
print("\nHeight of Student S4:", height_s4)

# Selecting chemistry grade of S3
chemistry_grade_s3 = grades_df.loc['S3', 'Chemistry']
print("\nChemistry Grade of S3:", chemistry_grade_s3)

# Selecting revenue for first 6 months and November
revenue_series = pd.Series(np.random.randint(5000, 20000, size=12), 
                           index=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
first_six_months = revenue_series.iloc[:6]
november_revenue = revenue_series.loc['Nov']
print("\nFirst Six Months Revenue:\n", first_six_months)
print("\nNovember Revenue:", november_revenue)
```

```python
# Lab Task 4: Understand and Perform Ufuncs

# Square root of 5 random integers
random_ints = np.random.randint(0, 10, 5)
sqrt_values = np.sqrt(random_ints)
print("\nSquare Root of Random Integers:\n", sqrt_values)

# Exponential of 3 random floating numbers
random_floats = np.random.uniform(-1, 1, 3)
exp_values = np.exp(random_floats)
print("\nExponential Values:\n", exp_values)

# Adding two random integer arrays and multiplying by 5
array1 = np.random.randint(0, 10, 3)
array2 = np.random.randint(0, 10, 3)
added_arrays = np.add(array1, array2)
multiplied_arrays = np.multiply(added_arrays, 5)
print("\nAdded Arrays:\n", added_arrays)
print("\nMultiplied Arrays:\n", multiplied_arrays)

# Sorting an array
random_array = np.random.randint(0, 10, 6)
sorted_array = np.sort(random_array)
print("\nSorted Array:\n", sorted_array)

# Mean and Standard Deviation of an array
random_large_array = np.random.randint(0, 100, 10)
mean_value = np.mean(random_large_array)
std_dev = np.std(random_large_array)
print("\nMean of Array:", mean_value)
print("Standard Deviation of Array:", std_dev)
```

```python
# Home Task

# Identifying missing values
nan_df = pd.DataFrame(np.random.rand(5, 3))
nan_df.iloc[2, 1] = np.nan  # Setting a NaN value
missing_values = nan_df.isna()
print("\nMissing Values in DataFrame:\n", missing_values)

# Dropping rows with missing values
dropna_df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, np.nan],
    'C': [7, 8, 9],
    'D': [np.nan, 11, 12]
})
dropna_cleaned = dropna_df.dropna()
print("\nDataFrame after Dropping NaN Rows:\n", dropna_cleaned)

# Filling NaN values with column mean
fillna_df = pd.DataFrame({
    'Col1': [10, 15, np.nan, 20, 25, np.nan],
    'Col2': [5, 8, 12, np.nan, 15, 18]
})
fillna_df['Col1'] = fillna_df['Col1'].fillna(fillna_df['Col1'].mean())
print("\nDataFrame after Filling NaN with Column Mean:\n", fillna_df)
```
