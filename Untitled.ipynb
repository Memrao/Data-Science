```python
# Creating empty dictionaries
empty_dict = {}  # Pythonic way
empty_dict2 = dict()  # Less Pythonic but valid

# Defining a dictionary with key-value pairs
grades = {"Joel": 80, "Tim": 95}  

# Accessing dictionary values
joels_grade = grades["Joel"]  # Equals 80
print("Joel's Grade:", joels_grade)

# Handling missing keys with try-except
try:
    kates_grade = grades["Kate"]  # Will raise KeyError
except KeyError:
    print("No grade for Kate!")

# Checking if a key exists in the dictionary
joel_has_grade = "Joel" in grades  # True
kate_has_grade = "Kate" in grades  # False
print("Joel in grades:", joel_has_grade)
print("Kate in grades:", kate_has_grade)

# Using `get()` to avoid KeyError
joels_grade = grades.get("Joel", 0)  # Returns 80
kates_grade = grades.get("Kate", 0)  # Returns 0
no_ones_grade = grades.get("No One")  # Returns None
print("Joel's Grade:", joels_grade)
print("Kate's Grade:", kates_grade)
print("No One's Grade:", no_ones_grade)

# Modifying and adding new values to the dictionary
grades["Tim"] = 99  # Updates Tim's grade
grades["Kate"] = 100  # Adds a new entry
num_students = len(grades)  # Total number of students
print("Updated Grades:", grades)
print("Total Students:", num_students)

# Dictionary representing structured data
tweet = {
    "user": "joelgrus",
    "text": "Data Science is Awesome",
    "retweet_count": 100,
    "hashtags": ["#data", "#science", "#datascience", "#awesome", "#yolo"]
}

# Getting keys, values, and key-value pairs
tweet_keys = tweet.keys()  # List of keys
tweet_values = tweet.values()  # List of values
tweet_items = tweet.items()  # List of (key, value) tuples

# Checking for key existence
print("Keys in tweet:", tweet_keys)
print("Values in tweet:", tweet_values)
print("Items in tweet:", tweet_items)

# Faster way to check for a key
print("User key exists:", "user" in tweet)  # True
print("joelgrus in tweet values:", "joelgrus" in tweet_values)  # True

# Using defaultdict for counting words
from collections import defaultdict

# Word counting using a normal dictionary
word_counts = {}
document = ["hello", "world", "hello", "python"]  # Example document

for word in document:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

print("Word Counts (Normal Dict):", word_counts)

# Using try-except for missing keys
word_counts = {}
for word in document:
    try:
        word_counts[word] += 1
    except KeyError:
        word_counts[word] = 1

print("Word Counts (Try-Except):", word_counts)

# Using get() method
word_counts = {}
for word in document:
    previous_count = word_counts.get(word, 0)
    word_counts[word] = previous_count + 1

print("Word Counts (Using get()):", word_counts)

# Using defaultdict for automatic handling of missing keys
word_counts = defaultdict(int)  # Default value is 0

for word in document:
    word_counts[word] += 1

print("Word Counts (Using defaultdict):", dict(word_counts))

# Using defaultdict with list, dict, and lambda functions
dd_list = defaultdict(list)  # Default value is an empty list
dd_list[2].append(1)  # {2: [1]}

dd_dict = defaultdict(dict)  # Default value is an empty dictionary
dd_dict["Joel"]["City"] = "Seattle"  # { "Joel": { "City": "Seattle" } }

dd_pair = defaultdict(lambda: [0, 0])  # Default value is [0,0]
dd_pair[2][1] = 1  # {2: [0,1]}

print("Defaultdict with list:", dict(dd_list))
print("Defaultdict with dict:", dict(dd_dict))
print("Defaultdict with lambda:", dict(dd_pair))
```

```python
# Generator function (lazy range)
def lazy_range(n):
    """A lazy version of range"""
    i = 0
    while i < n:
        yield i
        i += 1

# Consuming the generator
def do_something_with(i):
    print(i, end=" ")  # Example function to process values

for i in lazy_range(10):
    do_something_with(i)

print("\n")  # Newline for better readability

# Infinite generator
def natural_numbers():
    """Returns 1, 2, 3, ..."""
    n = 1
    while True:
        yield n
        n += 1

# Example usage with a break condition (avoiding infinite loop)
gen = natural_numbers()
for _ in range(10):  # Print first 10 natural numbers
    print(next(gen), end=" ")

print("\n")  # Newline for better readability

# Generator comprehension (lazy evaluation)
lazy_evens_below_20 = (i for i in lazy_range(20) if i % 2 == 0)
print("Lazy evens below 20:", list(lazy_evens_below_20))  # Converting to list to print

# Dictionary iteration (Python 3)
word_counts = {"hello": 5, "world": 10}
for key, value in word_counts.items():  # In Python 3, items() is used instead of iteritems()
    print(f"{key}: {value}")
```

```python
# By convention, we give classes PascalCase names
class Set:
    # These are the member functions
    # Every one takes a first parameter "self" (another convention)
    # that refers to the particular Set object being used

    def __init__(self, values=None):
        """This is the constructor.
        It gets called when you create a new Set.
        You would use it like:
        s1 = Set()  # empty set
        s2 = Set([1,2,2,3])  # initialize with values"""
        
        self.dict = {}  # Each instance of Set has its own dict property
                        # which is what we'll use to track memberships

        if values is not None:
            for value in values:
                self.add(value)

    def __repr__(self):
        """This is the string representation of a Set object.
        If you type it at the Python prompt or pass it to str(), it will return this."""
        return "Set: " + str(list(self.dict.keys()))

    # We'll represent membership by being a key in self.dict with value True
    def add(self, value):
        """Adds an element to the set."""
        self.dict[value] = True

    # Value is in the Set if it's a key in the dictionary
    def contains(self, value):
        """Checks if an element is in the set."""
        return value in self.dict

    def remove(self, value):
        """Removes an element from the set."""
        if value in self.dict:
            del self.dict[value]
        else:
            raise KeyError(f"{value} not found in the set")


# Example usage
s = Set([1, 2, 3])
s.add(4)
print(s.contains(4))  # True
s.remove(3)
print(s.contains(3))  # False
```

```python
from functools import partial, reduce

# Exponentiation function
def exp(base, power):
    return base ** power

# Using partial to create a function of one variable
two_to_the = partial(exp, 2)  # Now a function of one variable
print(two_to_the(3))  # 8

# Using partial to fill in named arguments
square_of = partial(exp, power=2)
print(square_of(3))  # 9

# Function to double a number
def double(x):
    return 2 * x

xs = [1, 2, 3, 4]

# List comprehension for doubling
twice_xs = [double(x) for x in xs]  # [2, 4, 6, 8]

# Using map for doubling
twice_xs = list(map(double, xs))  # Convert map object to list
print(twice_xs)  # [2, 4, 6, 8]

# Function that doubles a list using partial
list_doubler = partial(map, double)
twice_xs = list(list_doubler(xs))  # Convert map object to list
print(twice_xs)  # [2, 4, 6, 8]

# Function to multiply two numbers
def multiply(x, y):
    return x * y

# Using map with multiple lists
products = list(map(multiply, [1, 2], [4, 5]))  # Convert map object to list
print(products)  # [1 * 4, 2 * 5] = [4, 10]

# Function to check if a number is even
def is_even(x):
    """True if x is even, False if x is odd"""
    return x % 2 == 0

# List comprehension for filtering evens
x_evens = [x for x in xs if is_even(x)]  # [2, 4]
print(x_evens)

# Using filter for even numbers
x_evens = list(filter(is_even, xs))  # Convert filter object to list
print(x_evens)  # [2, 4]

# Function that filters a list using partial
list_evener = partial(filter, is_even)
x_evens = list(list_evener(xs))  # Convert filter object to list
print(x_evens)  # [2, 4]

# Using reduce to multiply all elements in a list
x_product = reduce(multiply, xs)  # = 1 * 2 * 3 * 4 = 24
print(x_product)

# Function that reduces a list using partial
list_product = partial(reduce, multiply)
x_product = list_product(xs)  # Again = 24
print(x_product)
```

```python
# Sample list of documents
documents = ["doc1", "doc2", "doc3", "doc4"]

# Example function to process documents
def do_something(index, document=None):
    if document is not None:
        print(f"Processing document {index}: {document}")
    else:
        print(f"Processing document index {index}")

# Pythonic way to iterate with index and document
for i, document in enumerate(documents):
    do_something(i, document)

# Pythonic way if you only need indexes
for i, _ in enumerate(documents):
    do_something(i)
```

```python
# Using zip to combine two lists
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]

# Converting zip object to a list
zipped = list(zip(list1, list2))  
print(zipped)  # Output: [('a', 1), ('b', 2), ('c', 3)]

# If the lists have different lengths, zip stops at the shortest list
list3 = [10, 20]
zipped_short = list(zip(list1, list3))  
print(zipped_short)  # Output: [('a', 10), ('b', 20)]

# Unzipping a list of tuples
pairs = [('a', 1), ('b', 2), ('c', 3)]
letters, numbers = zip(*pairs)  

print(letters)  # Output: ('a', 'b', 'c')
print(numbers)  # Output: (1, 2, 3)

# Using argument unpacking with functions
def add(a, b):
    return a + b

# Normal function call
print(add(1, 2))  # Output: 3

# Incorrect way: passing a list directly
# print(add([1, 2]))  # This would raise a TypeError

# Correct way: using argument unpacking (*)
print(add(*[1, 2]))  # Output: 3
```

```python
# Function decorator to double the output of any function
def doubler(f):
    def g(x):
        return 2 * f(x)
    return g

# Example function with a single argument
def f1(x):
    return x + 1

g = doubler(f1)
print(g(3))  # Output: 8 ( (3 + 1) * 2 )
print(g(-1)) # Output: 0 ( (-1 + 1) * 2 )

# Problem: This won't work with multiple arguments
def f2(x, y):
    return x + y

# g = doubler(f2)  # This will cause an error because g expects only 1 argument
# print(g(1, 2))  # TypeError

# Fixing the problem with *args and **kwargs
def magic(*args, **kwargs):
    print("Unnamed args:", args)
    print("Keyword args:", kwargs)

magic(1, 2, key="word", key2="word2")
# Output:
# Unnamed args: (1, 2)
# Keyword args: {'key': 'word', 'key2': 'word2'}

# Using argument unpacking to pass lists and dictionaries
def other_way_magic(x, y, z):
    return x + y + z

x_y_list = [1, 2]
z_dict = {"z": 3}

print(other_way_magic(*x_y_list, **z_dict))  # Output: 6

# Corrected doubler function that supports multiple arguments
def doubler_correct(f):
    """Works no matter what kind of inputs f expects"""
    def g(*args, **kwargs):
        """Whatever arguments g is supplied, pass them through to f"""
        return 2 * f(*args, **kwargs)
    return g

g = doubler_correct(f2)
print(g(1, 2))  # Output: 6 ( (1 + 2) * 2 )
```
